```python
from scipy.special import softmax

def score(example):
    # Tokenize the input text
    encoded_text = tokenizer(example, return_tensors='pt')

    # Perform a forward pass through the model to get the logits
    output = model(**encoded_text)

    # Extract the scores (logits) for the first text sample (batch size = 1)
    scores = output[0][0].detach().numpy()

    # Print the raw scores before softmax
    print("Raw scores before softmax:", scores)

    # Apply the softmax to convert logits to probabilities
    scores = softmax(scores)

    # Print the scores after softmax to check normalization
    print("Scores after softmax:", scores)

    # Define the labels for sentiment
    labels = ['Negative', 'Neutral', 'Positive']

    # Create a dictionary with the labels and corresponding scores
    scores_dict = {label: float(score) for label, score in zip(labels, scores)}

    # Print the final scores dictionary
    print("Scores dictionary:", scores_dict)

    return scores_dict
```

```python
from tqdm.notebook import tqdm

def scores(text):
    output = model(**encoded_text)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)

result = {}

for i, row in tqdm(df.iterrows(), total=len(df)):
  try:
    text = row['cleaned_comment']
    result[i] = scores(text)
  except RuntimeError:
    print(f'Broke for index {i}')
```

```python
for i, row in tqdm(df.iterrows(), total=len(df)):
    try:
        text = row['cleaned_comment']
        result[i] = scores(text)
    except ValueError as ve:
        print(f'ValueError for index {i}: {ve}')
    except RuntimeError:
        print(f'Broke for index {i}')
```

```python
results_df = pd.DataFrame.from_dict(result, orient='index')
results_df
```

```python
results_df.columns = ['negative', 'neutral', 'positive']
results_df
```

```python
for i, row in tqdm(df.iterrows(), total=len(df)):
    try:
        text = row['cleaned_comment']
        if isinstance(text, str) and text.strip():
            score_dict = scores(text)
            result[i] = {
                "comment": text,
                "negative": score_dict[0],  # if your function returns keys 0,1,2
                "neutral": score_dict[1],
                "positive": score_dict[2]
            }
        else:
            result[i] = {
                "comment": text,
                "negative": None,
                "neutral": None,
                "positive": None
            }
    except ValueError as ve:
        print(f'ValueError for index {i}: {ve}')
        result[i] = {
            "comment": None,
            "negative": None,
            "neutral": None,
            "positive": None
        }
    except RuntimeError:
        print(f'Broke for index {i}')
        result[i] = {
            "comment": None,
            "negative": None,
            "neutral": None,
            "positive": None
        }
```

```python
results_df = pd.DataFrame.from_dict(result, orient='index')
results_df
```
